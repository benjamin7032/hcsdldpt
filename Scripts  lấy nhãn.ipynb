```python
import os


objects = os.listdir()
folders = []
for obj in objects:
    if '.' not in obj:
        folders.append(obj)


# Danh sách tên thư mục con

# List để lưu tên tất cả các file
all_files = []
labels = []

# Duyệt qua từng thư mục con
for subfolder in folders:
    # Đường dẫn đến thư mục con
    subfolder_path = os.path.join(subfolder)
    # Kiểm tra nếu thư mục tồn tại
    if os.path.exists(subfolder_path):
        # Lấy danh sách các tệp tin trong thư mục con
        files = os.listdir(subfolder_path)
        # Duyệt qua từng tệp tin và thêm vào danh sách tên tất cả các file
        for file in files:
            # Đường dẫn đến tệp tin
            file_path = os.path.join(subfolder_path, file)
            # Kiểm tra nếu là tệp tin và không phải là thư mục
            if os.path.isfile(file_path):
                all_files.append(file)
                labels.append(subfolder)

# In ra danh sách tên tất cả các file
print("Danh sách tên tất cả các file:")
for file in all_files:
    print(file)
print("Danh sách tên tất cả các nhãn:")
for label in labels:
    print(label)
```

```python
import pandas as pd
df = pd.DataFrame({'File': all_files, 'label': labels})
df
```

```python
df.to_csv('data.csv', index= False)
```
